1.) Age: This refers to the age of the individual in years. Age is a significant factor in assessing cardiovascular health as certain heart conditions become more prevalent with age.

2.) Sex: This refers to the biological sex of the individual and is usually represented as binary data, with 0 typically indicating female and 1 indicating male. Sex can influence the prevalence and manifestation of various heart conditions.

3.) CP (Chest Pain): This is a categorical variable indicating the type of chest pain experienced by the individual. Common categories include typical angina, atypical angina, non-anginal pain, and asymptomatic. Different types of chest pain can indicate different underlying heart conditions.

4.) Trestbps (Resting Blood Pressure): This represents the resting blood pressure of the individual measured in millimeters of mercury (mmHg). Blood pressure is an important indicator of heart health, with high blood pressure (hypertension) being a risk factor for various cardiovascular diseases.

5.) Chol (Serum Cholesterol Level): This refers to the level of cholesterol in the bloodstream, typically measured in milligrams per deciliter (mg/dL). High levels of cholesterol, particularly low-density lipoprotein (LDL) cholesterol, are associated with an increased risk of heart disease.

6.) FBS (Fasting Blood Sugar): This represents the level of glucose (sugar) in the bloodstream after a period of fasting, usually measured in milligrams per deciliter (mg/dL). Elevated fasting blood sugar levels can be indicative of diabetes, which is a risk factor for heart disease.

7.) Restecg (Resting Electrocardiographic Results): This is a categorical variable indicating the results of the resting electrocardiogram (ECG or EKG), which measures the electrical activity of the heart. The results may include normal, ST-T wave abnormality, and probable or definite left ventricular hypertrophy.

8.) Thalach (Maximum Heart Rate Achieved): This represents the maximum heart rate achieved during exercise, typically measured in beats per minute (bpm). Maximum heart rate is often used to assess cardiovascular fitness and can help determine appropriate exercise intensity.

9.) Exang (Exercise Induced Angina): This is a binary variable indicating whether the individual experienced exercise-induced angina during cardiac stress testing. Angina is chest pain or discomfort caused by reduced blood flow to the heart muscle, typically due to coronary artery disease.

10.) Oldpeak (ST Depression Induced by Exercise Relative to Rest): This refers to the amount of ST segment depression observed on the ECG during exercise stress testing, relative to rest. ST depression can indicate myocardial ischemia, which occurs when the heart muscle is not receiving enough oxygen-rich blood. Oldpeak is measured in millimeters (mm).

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, roc_curve, roc_auc_score
```

Data Collection and Processing

```python
# loading the csv data to a Pandas DataFrame
heart_data = pd.read_csv('Dataset.csv')
```

1 --> Defective Heart

0 --> Healthy Heart

Splitting the Features and Target

```python
X = heart_data.drop(columns='target', axis=1)
Y = heart_data['target']
```

```python
print(X)
```

```python
print(Y)
```

Splitting the Data into Training data & Test Data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Model Training

Logistic Regression

```python
model = LogisticRegression()
```

```python
# training the LogisticRegression model with Training data
model.fit(X_train, Y_train)
```

Model Evaluation

Accuracy Score

```python
# accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy on Training data : ', training_data_accuracy)
```

```python
# accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy on Test data : ', test_data_accuracy)
```

```python
# Confusion Matrix
conf_matrix = confusion_matrix(Y_test, X_test_prediction)
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, cmap="YlGnBu", fmt='g')
plt.xlabel('Predicted labels')
plt.ylabel('True labels')
plt.title('Confusion Matrix')
plt.show()
```

```python
# ROC Curve
y_pred_prob = model.predict_proba(X_test)[:,1]
fpr, tpr, thresholds = roc_curve(Y_test, y_pred_prob)
plt.figure(figsize=(8, 6))
plt.plot([0, 1], [0, 1], 'k--')
plt.plot(fpr, tpr, label='Logistic Regression')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.show()
```

```python
# Additional Predictions using Seaborn
sns.pairplot(heart_data, hue='target', vars=['age', 'trestbps', 'chol', 'thalach'])
plt.show()
```
